## Titanic dataset

Using Titanic dataset to **Classify** whether or not a person survived or not.

#### Importing necessary libraries, methods...

````python
import pandas as pd
import numpy as np
from pprint import pprint
````

#### Loading Titanic dataset

````python
titanicfile_path = r'C:\Users\ANIMESH\Documents\Python Scripts\datasets\train.csv'

titanic = pd.read_csv (titanicfile_path)

print (titanic.info ())
titanic.head ()
````

#### Dropping unnecessary columns

````python
print (titanic.columns, len (titanic.columns))

titanic.drop ('PassengerId', axis = 1, inplace = True)
titanic.drop ('Ticket', axis = 1, inplace = True)
titanic.drop ('Cabin', axis = 1, inplace = True)

print (titanic.columns, len (titanic.columns))

titanic.head ()
````

````python
print (titanic.info ())
````

#### Finding median age on the basis of sex

````python
print (titanic.groupby ('Sex')['Age'].median ())
# print (titanic.groupby ('Sex')['Age'].quantile (0.5)) # does same work
# help (titanic.groupby ('Sex')['Age'])
````

#### Seeing which rows have NaN values

````python
# print (titanic['Age'].isnull ().sum ())
titanic[titanic['Age'].isnull ()]
````

#### Creating two new rows

Surname: to groupby surname later

Salutation: to assign appropriate age (median age) to the missing value rows

````python
titanic['Surname'] =  titanic['Name'].apply (lambda x: x.split(',')[0])
titanic['Salutation'] = titanic['Name'].apply (lambda x: x.split(',')[-1].split('.')[0][1:])

titanic.head ()
````

#### Reducing the number of unique salutations in the 'Surname' column

````python
print (titanic['Salutation'].unique (), len (titanic['Surname'].unique ()))
# print (len (titanic['Surname'].unique ()))
print (titanic.groupby ('Surname')['Survived'].sum ())
````

````python
Mr = ['Don', 'Rev', 'Major', 'Sir', 'Col', 'Capt', 'Jonkheer']
Mrs = ['Mme', 'Lady', 'the Countess']
Miss = ['Ms', 'Mlle']

for i in range (len (titanic)):
    if titanic['Salutation'][i] in Mr:
        titanic.loc[i, 'Salutation'] = 'Mr'
    elif titanic['Salutation'][i] in Mrs:
        titanic.loc[i, 'Salutation'] = 'Mrs'
    elif titanic['Salutation'][i] in Miss:
        titanic.loc[i, 'Salutation'] = 'Miss'
    elif titanic['Salutation'][i] == 'Dr':
        if titanic['Sex'][i] == 'male':
            titanic.loc[i, 'Salutation'] = 'Mr'
        else:
            titanic.loc[i, 'Salutation'] = 'Mrs'
                
print (titanic['Salutation'].unique ())
print (titanic.info ())
````

#### Seeing median age of the dataset based on sex and salutation

*Note*: NaN values are still in the dataset, but still couldn't hurt getting some insight from the available data, **right**?

````python
print (titanic.groupby (['Sex', 'Salutation'])['Age'].median ())
print (type (titanic.groupby (['Sex', 'Salutation'])['Age'].median ()))
````

#### Filling the 'Embarked' column

````python
import random

allowed = ['S', 'C', 'Q']

for i in range (len (titanic)):
    if titanic['Embarked'][i] not in allowed:
        titanic.loc[i, 'Embarked'] = random.choice (allowed)

print (titanic.info ())
````

#### Filling the missing values based on previous insight

*Of median age based on the gender and age group (based on salutation)*

````python
for i in range (len (titanic)):
    if np.isnan (titanic['Age'][i]):
        if titanic['Salutation'][i] == 'Mr':
            titanic.loc[i, 'Age'] = 30
        elif titanic['Salutation'][i] == 'Master':
            titanic.loc[i, 'Age'] = 3.5
        elif titanic['Salutation'][i] == 'Mrs':
            titanic.loc[i, 'Age'] = 35
        elif titanic['Salutation'][i] == 'Miss':
            titanic.loc[i, 'Age'] = 21
            
print (titanic.info ())
````

### Cleaning done.

````python
from sklearn.preprocessing import LabelEncoder

cat_mask = (titanic.dtypes == object)

cat_columns = titanic.columns[cat_mask].tolist ()

le = LabelEncoder ()

titanic[cat_columns] = titanic[cat_columns].apply (lambda x: le.fit_transform (x))

titanic.head (10)
````

````python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder (categorical_features = cat_mask, sparse = False)

titanic_encoded = ohe.fit_transform (titanic)

print (pd.DataFrame(titanic_encoded)[:10])
print (titanic_encoded.shape)

# print (titanic_encoded[:10])
````

````python
import xgboost as xgb
from sklearn.model_selection import train_test_split

xgbcl = xgb.XGBClassifier (objective = 'binary:logistic', n_estimators = 100, seed = 15)

X = titanic_encoded[:, 1:]
Y = titanic_encoded[:, 0]

pprint (X.shape)
pprint (Y.shape)

X_train, X_test, Y_train, Y_test = train_test_split (X, Y, test_size = 0.3, random_state = 15)

xgbcl.fit (X_train, Y_train)

preds = xgbcl.predict(X_test)

accuracy = np.sum(preds==Y_test)/Y_test.shape[0]
print('Accuracy:', (accuracy))
# pprint (preds)
````
